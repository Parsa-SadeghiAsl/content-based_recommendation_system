# Importing Liberaries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

### Reading CSV File Into Dataframes:

```python
movies_df = pd.read_csv('movies.csv') # Movies Information 
ratings_df = pd.read_csv('ratings.csv') # User Ratings To Movies

ratings_df.head()
```

```python
movies_df.head()
```

## Preprocessing the data

Separating Years From Title Column and Store It In a New one

```python
# Extract year inside "()" using regex
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))', expand=False)
# Removing parentheses
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)', expand=False)
# Removing years from title and strip it to remove white spaces
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))', '')
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

spliting genres and making them lists

```python
# separating evry genre by '|'
movies_df['genres'] = movies_df.genres.str.split('|')  
movies_df.head()
```

Storing every genre of single movie in diffrent column that contains value of either 1 or 0 to present that if that movie have that genre or not

```python
movieswithgenres_df = movies_df.copy()

for index, row in movies_df.iterrows():
    for genre in row['genres']:
        movieswithgenres_df.at[index, genre] = 1
    
# replacing Nan values with 0 to show that movie doesn't have that genre
movieswithgenres_df = movieswithgenres_df.fillna(0)
movieswithgenres_df.head()
```

```python
movieswithgenres_df.head(50)
```

```python
# Remove timestamp from ratings dataframe
ratings_df = ratings_df.drop('timestamp', 1)
ratings_df.head()
```

## Content-based recommendation system

User input ratings 

```python
userinput = [
    {'title':'Casino', 'rating':5},
    {'title':'Assassins', 'rating':3.4},
    {'title':'Heat', 'rating':2},
    {'title':'Jumanji', 'rating':4.6},
    {'title':'Twelve Monkeys (a.k.a. 12 Monkeys)', 'rating':5}
]
# converting Dicts into dataframe with pandas
inputmovies = pd.DataFrame(userinput)
inputmovies
```

Finding Movieid and add them to inputmovies data frame

```python
inputid = movies_df[movies_df['title'].isin(inputmovies['title'].tolist())]
inputmovies = pd.merge(inputid, inputmovies)
inputmovies = inputmovies.drop('genres', 1).drop('year', 1)

inputmovies
```

Finding user watched movies from movies with genres and making them a new data frame

```python
usermovies = movieswithgenres_df[movieswithgenres_df['movieId'].isin(inputmovies['movieId'].tolist())]
usermovies
```

```python
# Resettng the index
usermovies = usermovies.reset_index(drop=True)
# We just need genres table so we remove other columns 
usergenretable = usermovies.drop('movieId', 1).drop('year', 1).drop('genres', 1).drop('title', 1)
usergenretable
```

### User profile

```python
# multiply to get weights for user profile
userprofile = usergenretable.transpose().dot(inputmovies['rating'])
userprofile
```

```python
genretable = movieswithgenres_df.set_index(movieswithgenres_df['movieId'])
genretable = genretable.drop('movieId', 1).drop('year', 1).drop('genres', 1).drop('title', 1)
genretable.head()
```

Multiply genres from main data frame by weights to get weighted avg 

```python
recommendationtable_df = ((genretable*userprofile).sum(axis=1))/(userprofile.sum())
```

```python
# sorting recommends
recommendationtable_df = recommendationtable_df.sort_values(ascending=False)

recommendationtable_df.head()
```
